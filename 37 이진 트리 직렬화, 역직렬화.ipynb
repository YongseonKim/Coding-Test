```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        # bfs로 구현하자.
        queue = collections.deque([root])
        lst =[]
        while queue:
            cur_node = queue.popleft()
            if cur_node is None:
                lst.append('-')
                continue
            lst.append(str(cur_node.val))
            # if cur_node.left:
            queue.append(cur_node.left)
            # if cur_node.right:
            queue.append(cur_node.right)

        # 뒤에 '-'는 어떻게 제거?
        # 역질렬화 하면서 제거된다.
        # 직렬화는 투리를 배열로 바꿔주는 것 !!
        # 배열 형태로 표현되어야한다.
        # print(lst)
        return ' '.join(lst)
        
        

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        # 배열을 입력받아 Tree로 만드는 것
        # Queue 형태로 만들어갈 것
        if data == '-':
            return None
        nodes = data.split()
        root = TreeNode(int(nodes[0]))
        
        queue = collections.deque([root])
        index =1 # pointer
        
        while queue:
            node = queue.popleft()
            # 왼쪽 
            if nodes[index] != '-':
                node.left = TreeNode(int(nodes[index])) 
                queue.append(node.left)
            index +=1
            
            #오른쪽
            if nodes[index] != '-':
                node.right = TreeNode(int(nodes[index]))
                queue.append(node.right)        
            index +=1
            
        return root
        

# Your Codec object will be instantiated and called as such:
# codec = Codec()
# codec.deserialize(codec.serialize(root))
```
