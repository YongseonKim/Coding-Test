```python
class Solution:
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        # col, row 
        
        if not matrix:
            return  False
        # 오른쪽 끝에서 시작해서 내려가는 것
        row = 0
        col = len(matrix[0])-1
        
        while row <=len(matrix)-1 and col >=0:
            if matrix[row][col] == target:
                return True
            # target이 크면 아래로 이동 
            elif matrix[row][col] <target:
                row+=1
            # target이 작으면 왼쪽으로 이동
            elif matrix[row][col]>target:
                col -=1
                
        return False
                
                
```
