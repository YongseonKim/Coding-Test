```python
# 퀵Sort
'''
퀵 정렬 (Quick Sort) 파이썬으로 구현
퀵 정렬은 분할 정복 알고리즘의 좋은 예이다.

리스트 중 하나를 pivot으로 정하고, pivot보다 작은 아이템은 왼쪽, pivot보다 큰 아이템은 오른쪽으로 보내면서 pivot의 자리를 확정한다.

그리고 왼쪽 리스트와 오른쪽 리스트에 대해서 재귀형식으로 진행.

평균 시간복잡도는 O(n*log n), 최악의 시간복잡도는 O(n^2)
https://debuglog.tistory.com/67?category=724430
'''

def quicksort(lst):
    if len(lst)>1:  
        pivot = lst[-1]
        left,mid, right = [],[],[]
        for i in range(len(lst)-1):
            if lst[i]<pivot : left.append(lst[i])
            elif  lst[i]>pivot : right.append(lst[i])
            else : mid.append(lst[i])
        mid.append(pivot)
        return quicksort(left) + mid + quicksort(right)
    
    else : return lst
            

lst= [3,4,5,6,1,2,3,2,1,3,5,6,8,90,6,3,3]
quicksort(lst)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the makeAnagram function below.
def makeAnagram(a, b):
    # count해서 처리하면 되는데, set을 잘 이용하면 된다. 
    # list.count() 함수도 사용할 수 있다. 
    # 다 왔는데 마지막에 마무리가 안되엇다. 
    # count를 한것은 맞았는데 그 다음에 count 한것을 기준으로 돌려서 잇는지 없는지를 체크했어야한다.
    # 그 두개의 차이를 계산하면 되는 것이였다.
    
    
    dict1 =  {x:0 for x in list(set(a))}
    dict2 =  {x:0 for x in list(set(b))}

    #count
    # 중복하여 나오는 횟수가 다른 것들도 지워져야하는데 
    # 카운트를 세서 dict로 저장해둔 다음에 
    # 양 쪽의 카운트를 반복문을 돌려가며 차이를 구하고 저장하는 것
    for x in a : dict1[x]+=1
    for x in b : dict2[x]+=1

    cnt = 0
    for key in dict1.keys():
        if key in dict2 :
            cnt += abs(dict1[key]-dict2[key])
        else: cnt += dict1[key]

    for key in dict2.keys():
        if key in dict1 : continue
        else : cnt += dict2[key]
    return cnt 


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = input()

    b = input()

    res = makeAnagram(a, b)

    fptr.write(str(res) + '\n')

    fptr.close()
```

```python
# https://programmers.co.kr/learn/courses/30/lessons/43162
def dfs(computers,start,nodes):
    global answer 
    
    #start point를 기점으로 계속 이어나가보자.
    stack = [start] # 시작지점을 stack에 넣는다.
    while stack : # stack이 빌 때까지 반복한다.
        k = stack.pop() #마지막에 들어간 것을 빼서 연결할 곳을 찾자
        if nodes[k]==0 : nodes[k]=1 # 본걸로 바꾸는 것
        
        for idx in range(len(computers)): #반복하면서 찾는것
            if nodes[idx] ==0  and computers[k][idx] ==1 : # 아직 방문하지 않았고 연결되어있다면
                stack.append(idx) # stack에 연결하겠다.
        print(stack, start)
        
    return

def solution(n, computers):
    answer = 0
    nodes = [0]*n
    
    i =0

    while 0 in nodes : # 모든 데이터를 볼때까지 반복할 것
        if nodes[i]==0: # 아직 방문안했다면,
            dfs(computers, i,nodes)
            answer +=1
        i +=1            
        
    return answer

#  dfs 로 풀어야 하는 문제
 # searching을 어떻게 할 것인지.
    
```
