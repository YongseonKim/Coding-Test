```python
# Set을 사용하면 속도가 더 빠르다 
class Solution:
    def findMinHeightTrees(self, n: int, edges: List[List[int]]) -> List[int]:
        
        if n<=1:
            return [0]
        
        # 그래프의 중간에 있는 노드를 찾아 그래프를 만들면
        # 최소 높이의 그래프가 된다.
        # 중간에 있는 노드는 어떻게 찾을래- leaf 노드를 계속 지워 가는 것
        
        # Graph 구성
        graph = collections.defaultdict(list)
        
        # 양방향 그래프를 구축 해야한다.
        for i, j in edges:
            graph[i].append(j)
            graph[j].append(i)
            
    
        # print(graph)
        # leaf node 판단은 graph value가 1인것들을 의미한다.
        # Leaf를 찾아서 Leaf를 포함한 것들을 다 지우는 것
        # n을 이용해야 한다.
        
        # leaves 찾기
        leaves = set()
        for i in range(n):
            if len(graph[i])==1 :
                leaves.add(i)
        
        while n>2 : # 홀수 개수이면 1개, 짝수 개수이면 2개가 남을 것
            n-=len(leaves)
            # leaf 노드 지우기
            new_leaves = set() ## 속도 올리기
            for leaf in leaves: # stack으로 처리는 못하고 Queue로 할 수 있지 않나 
                contact = graph[leaf].pop() # graph 에 있는 값 가져오기
                graph[contact].remove(leaf)
                
                if len(graph[contact]) ==1 : # 변경 사항이 생긴 주변애가 Leaf인지 판단.
                    new_leaves.add(contact)
            leaves = new_leaves
            
        # 마지막에 남은 leaves return
        return leaves
```
